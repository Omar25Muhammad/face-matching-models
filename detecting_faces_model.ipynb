```python
from ultralytics import YOLO
import cv2
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# load a pretrained YOLOv8n detection model
model = YOLO('/content/drive/MyDrive/Colab Notebooks/ard/yolov8_model/yolov8n-face.pt')
```

```python
image = ''
detections = model()
```

```python
boxes = detections[0].boxes.data.squeeze().tolist()
orig_img = detections[0].orig_img
```

```python
# Extract bounding box coordinates
x1, y1, x2, y2 = map(int, boxes[:4])

# Draw bounding box on the image
cv2.rectangle(orig_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Display the image with the bounding box
from google.colab.patches import cv2_imshow
cv2_imshow(orig_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
